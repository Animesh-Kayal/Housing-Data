Separating Year from date

```python
def year(value):
    return value.split()[-1]

data['year_sold']= data['Date House was Sold'].map(year)
data['year_sold'].head(5)
```

Treating Outliers

```python
import seaborn as sns
```

```python
sns.boxplot(x = data['Sale Price'])
```

```python
q1 = data['Sale Price'].quantile(0.25)
q3 = data['Sale Price'].quantile(0.75)
q1,q3
```

```python
iqr = q3-q1
iqr
```

```python
upper_limit = q3 + iqr*0.5
lower_limit = q1 + iqr*0.5
lower_limit, upper_limit
```

```python
def limit_imputer(value):
    if value > upper_limit:
        return upper_limit
    if value < lower_limit:
        return lower_limit
    else:
        return value
```

```python
data['Sale Price'] = data['Sale Price'].apply(limit_imputer)
data['Sale Price'].describe()
```

Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = X
 #Calculating VIF for every column

VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range (vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
VIF[VIF == VIF.max()].index[0]
```

```python
def MC_remover(data):
     vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index= data.columns)
     if vif.max() > 5:
        print(vif[vif == vif.max()]. index[0], "has been removed")
        data = data.drop(columns=[vif[vif == vif.max()].index[0]])
        return data
     else:
        print("No Multicollinearity present anymore")

        return data
```

```python
for i in range(5):
    vif_data = MC_remover(vif_data)

vif_data.head()
```

```python
VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index= vif_data.columns)
VIF, len(vif_data.columns)
```

Standardisation

```python
from sklearn import preprocessing
scale = preprocessing.StandardScaler()
```

```python
X_train = scale.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scale.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
predictions = model.predict(X_test)
model.score(X_test, y_test)
```

Residuals

```python
residuals = predictions - y_test
residual_table = pd.DataFrame({'residuals': residuals, 'predictions': predictions})
residual_table = residual_table.sort_values(by = 'predictions')
```

```python
z = [i for i in range(int(residual_table['predictions'].max()))]
k = [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
from turtle import color


plt.figure(dpi = 150, figsize=(18, 9))
plt.scatter(residual_table['predictions'], residual_table['residuals'], color= 'blue', s=2)
plt.plot(z, k, color= 'red', linewidth = 3.5, label= 'Regression line')
plt.ylim(-800000, 800000)
plt.xlabel('fitted points(ordered by prediction)')
plt.ylabel('residuals')
plt.title('Residual Plot')
plt.legend()
plt.show()
```
